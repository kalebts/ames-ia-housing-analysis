## sample workflow

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
train = pd.read_csv('../datasets/train.csv')
test = pd.read_csv('../datasets/test.csv')
```

```python
train.head()
```

```python
test.head()
```

```python
X = train[['Gr Liv Area']]
y = train['SalePrice']
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
X_test = test[['Gr Liv Area']]
preds = lr.predict(X_test)
```

```python
test['SalePrice'] = preds
```

```python
submission = test[['Id', 'SalePrice']]
```

```python
submission.shape
```

```python
submission.to_csv('basic_model.csv', index = False) # index = False is important for kaggle submission
```
